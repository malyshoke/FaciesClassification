Описание столбцов
1. __(GR)__: этот каротажный инструмент измеряет гамма-излучение из пласта. Хороший показатель для определения содержания сланца.<br>
2. __(ILD_log10)__: это измерение удельного сопротивления, которое применяется для определения содержания флюидов в пласте.<br>
3. __(PE)__: каротаж на основе фотоэлектрического эффекта может быть использован для определения литологии (минерального состава породы).<br>
4. __(DeltaPHI)__: Phi - это индекс пористости в петрофизике. Для измерения пористости существует несколько методов, таких как нейтронный и плотностной. Здесь приведены различия между ними.<br>
5. __(PNHIND)__: среднее значение нейтронного и плотностного каротажа.<br>
6. __(NM_M)__ :неморской морской индикатор <br>
7. __(RELPOS)__: относительное положение <br>

Девять дискретных фаций (классов пород) следующие:
1. __(SS)__   неморской песчаник
2. __(CSiS)__ Неморской грубый алевролит
3. __(FSiS)__ Неморской мелкозернистый алевролит
4. __(SiSH)__ Морские алевролиты и сланцы 
5. __(MS)__   Грязевик (известняк)
6. __(WS)__   Известняк (известняк)
7. __(D)__    Доломит
8. __(PS)__   Пакстоун-зернистый камень (известняк)
9. __(BS)__   Филлоидно-водорослевый бафлстоун (известняк)

Фации, их сокращенные метки и приблизительные соседи.

Фации |Метки| Соседние фации
:---: | :---: |:--:
1 |SS| 2
2 |CSiS| 1,3
3 |FSiS| 2
4 |SiSh| 5
5 |MS| 4,6
6 |WS| 5,7
7 |D| 6,8
8 |PS| 6,7,9
9 |BS| 7,8

```python
import matplotlib.pyplot as plt

def plotlog(dataframe, column_to_plot):
    fig, ax = plt.subplots(figsize=(8, 6))

    # Устанавливаем общие пределы для оси y
    ymin, ymax = dataframe.index.min(), dataframe.index.max()

    # Строим график для выбранного столбца
    ax.plot(dataframe[column_to_plot], dataframe.index)
    ax.set_ylim(ymax, ymin)  # Инвертируем ось y, чтобы глубина увеличивалась сверху вниз
    ax.set_ylabel('Depth')
    ax.set_xlabel(column_to_plot)
    ax.grid(True)
    ax.invert_yaxis()  # Инвертируем ось y
    ax.xaxis.set_tick_params(rotation=45)  # Поворачиваем метки оси x на 45 градусов

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.axes_grid1 import make_axes_locatable

# 1=sandstone  2=c_siltstone   3=f_siltstone 
# 4=marine_silt_shale 5=mudstone 6=wackestone 7=dolomite
# 8=packstone 9=bafflestone
facies_colors = ['#F4D03F', '#F5B041','#DC7633','#6E2C00',
       '#1B4F72','#2E86C1', '#AED6F1', '#A569BD', '#196F3D']

facies_labels = ['SS', 'CSiS', 'FSiS', 'SiSh', 'MS',
                 'WS', 'D','PS', 'BS']
#facies_color_map is a dictionary that maps facies labels
#to their respective colors
facies_color_map = {}
for ind, label in enumerate(facies_labels):
    facies_color_map[label] = facies_colors[ind]


def make_facies_log_plot(logs, facies_colors):
    #make sure logs are sorted by depth
    logs = logs.sort_values(by='Depth')
    cmap_facies = colors.ListedColormap(
            facies_colors[0:len(facies_colors)], 'indexed')
    
    ztop=logs.Depth.min(); zbot=logs.Depth.max()
    
    cluster=np.repeat(np.expand_dims(logs['Facies'].values,1), 100, 1)
    
    f, ax = plt.subplots(nrows=1, ncols=6, figsize=(8, 12))
    ax[0].plot(logs.GR, logs.Depth, '-g')
    ax[1].plot(logs.ILD_log10, logs.Depth, '-')
    ax[2].plot(logs.DeltaPHI, logs.Depth, '-', color='0.5')
    ax[3].plot(logs.PHIND, logs.Depth, '-', color='r')
    ax[4].plot(logs.PE, logs.Depth, '-', color='black')
    im=ax[5].imshow(cluster, interpolation='none', aspect='auto',
                    cmap=cmap_facies,vmin=1,vmax=9)
    
    divider = make_axes_locatable(ax[5])
    cax = divider.append_axes("right", size="20%", pad=0.05)
    cbar=plt.colorbar(im, cax=cax)
    cbar.set_label((17*' ').join([' SS ', 'CSiS', 'FSiS', 
                                'SiSh', ' MS ', ' WS ', ' D  ', 
                                ' PS ', ' BS ']))
    cbar.set_ticks(range(0,1)); cbar.set_ticklabels('')
    
    for i in range(len(ax)-1):
        ax[i].set_ylim(ztop,zbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)
    
    ax[0].set_xlabel("GR")
    ax[0].set_xlim(logs.GR.min(),logs.GR.max())
    ax[1].set_xlabel("ILD_log10")
    ax[1].set_xlim(logs.ILD_log10.min(),logs.ILD_log10.max())
    ax[2].set_xlabel("DeltaPHI")
    ax[2].set_xlim(logs.DeltaPHI.min(),logs.DeltaPHI.max())
    ax[3].set_xlabel("PHIND")
    ax[3].set_xlim(logs.PHIND.min(),logs.PHIND.max())
    ax[4].set_xlabel("PE")
    ax[4].set_xlim(logs.PE.min(),logs.PE.max())
    ax[5].set_xlabel('Facies')
    
    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]); ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[5].set_yticklabels([])
    ax[5].set_xticklabels([])
    f.suptitle('Well: %s'%logs.iloc[0]['Well Name'], fontsize=14,y=0.94)


def compare_facies_plot(logs, compadre, facies_colors):
    """plot the facies plot as a function of depth for both the prediction
       and the actual lithofacies labels.
    """
   
    #make sure logs are sorted by depth
    logs = logs.sort_values(by='Depth')
    cmap_facies = colors.ListedColormap(
            facies_colors[0:len(facies_colors)], 'indexed')
    
    ztop=logs.Depth.min(); zbot=logs.Depth.max()
    
    cluster1 = np.repeat(np.expand_dims(logs['Facies'].values,1), 100, 1)
    cluster2 = np.repeat(np.expand_dims(logs[compadre].values,1), 100, 1)
    
    f, ax = plt.subplots(nrows=1, ncols=7, figsize=(9, 12))
    ax[0].plot(logs.GR, logs.Depth, '-g')
    ax[1].plot(logs.ILD_log10, logs.Depth, '-')
    ax[2].plot(logs.DeltaPHI, logs.Depth, '-', color='0.5')
    ax[3].plot(logs.PHIND, logs.Depth, '-', color='r')
    ax[4].plot(logs.PE, logs.Depth, '-', color='black')
    im1 = ax[5].imshow(cluster1, interpolation='none', aspect='auto',
                    cmap=cmap_facies,vmin=1,vmax=9)
    im2 = ax[6].imshow(cluster2, interpolation='none', aspect='auto',
                    cmap=cmap_facies,vmin=1,vmax=9)
    
    divider = make_axes_locatable(ax[6])
    cax = divider.append_axes("right", size="20%", pad=0.05)
    cbar=plt.colorbar(im2, cax=cax)
    cbar.set_label((17*' ').join([' SS ', 'CSiS', 'FSiS', 
                                'SiSh', ' MS ', ' WS ', ' D  ', 
                                ' PS ', ' BS ']))
    cbar.set_ticks(range(0,1)); cbar.set_ticklabels('')
    
    for i in range(len(ax)-2):
        ax[i].set_ylim(ztop,zbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)
        
    ax[0].set_xlabel("GR")
    ax[0].set_xlim(logs.GR.min(),logs.GR.max())
    ax[1].set_xlabel("ILD_log10")
    ax[1].set_xlim(logs.ILD_log10.min(),logs.ILD_log10.max())
    ax[2].set_xlabel("DeltaPHI")
    ax[2].set_xlim(logs.DeltaPHI.min(),logs.DeltaPHI.max())
    ax[3].set_xlabel("PHIND")
    ax[3].set_xlim(logs.PHIND.min(),logs.PHIND.max())
    ax[4].set_xlabel("PE")
    ax[4].set_xlim(logs.PE.min(),logs.PE.max())
    ax[5].set_xlabel('Facies')
    ax[6].set_xlabel(compadre)
    
    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]); ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[5].set_yticklabels([])
    ax[5].set_xticklabels([])
    ax[6].set_xticklabels([])
    f.suptitle('Well: %s'%logs.iloc[0]['Well Name'], fontsize=14,y=0.94)
```

```python
X = df[['Depth', 'GR', 'ILD_log10',
       'DeltaPHI', 'PHIND', 'PE', 'NM_M', 'RELPOS']]
y = df['Facies']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    random_state=0, 
                                                    test_size=0.2)
```

```python
from sklearn.multiclass import OneVsRestClassifier ### Один против всех
from sklearn.linear_model import SGDClassifier ### 

SGDClassifier(loss='log') ### max(0; 1-M)

pipe = Pipeline([("scaler", StandardScaler()),
                 ("one_vs_all", OneVsRestClassifier(SGDClassifier()))])

pipe.fit(X_train, y_train)
```

```python
np.mean((pipe.predict(X_test) == y_test)), np.mean((pipe.predict(X_train) == y_train))
```

```python
### Переберем различные параметры базовых моделей

from sklearn.model_selection import GridSearchCV

param_grid = {
    'one_vs_all__estimator__loss': ['hinge', 'log', 'modified_huber',
                                    'squared_hinge', 'perceptron'],
    
    'one_vs_all__estimator__penalty': ['l1', 'l2', 'elasticnet'],
    
    'one_vs_all__estimator__alpha': [0.001, 0.01, 0.1, 0.5, 1, 10]
}

from sklearn.model_selection import PredefinedSplit

test_fold = [0 if x in X_train.index else -1 for x in X.index]
ps = PredefinedSplit(test_fold)

search = GridSearchCV(pipe, param_grid, cv=ps)

search.fit(X, y)

print(f"Best parameter (CV score={search.best_score_:.5f}):")
print(search.best_params_)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# Определяем параметры для поиска
param_grid = {
    'svm__kernel': ['linear', 'rbf', 'poly'],
    'svm__C': [0.1, 1, 10],
    'svm__gamma': ['scale', 'auto'],
}

# Создаем объект SVM
svm = SVC()

# Создаем конвейер для объединения стандартизации и SVM
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', svm)
])

# Создаем объект GridSearchCV
search = GridSearchCV(pipe, param_grid, cv=ps)

# Запускаем поиск
search.fit(X, y)

# Выводим лучшие параметры и оценку
print(f"Best parameter (CV score={search.best_score_:.5f}):")
print(search.best_params_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Создаем модель случайного леса
rf_classifier = RandomForestClassifier(random_state=42)

# Определяем параметры для поиска для случайного леса
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Создаем объект GridSearchCV для случайного леса
rf_search = GridSearchCV(rf_classifier, rf_param_grid, cv=5)

# Запускаем поиск для случайного леса
rf_search.fit(X_train, y_train)

# Выводим лучшие параметры и оценку для случайного леса
print("Best parameters for Random Forest:", rf_search.best_params_)
print("Accuracy of Random Forest:", rf_search.best_score_)
```

```python
# Создаем модель градиентного бустинга
gb_classifier = GradientBoostingClassifier(random_state=42)

# Определяем параметры для поиска для градиентного бустинга
gb_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.01, 0.001],
    'max_depth': [3, 5, 7]
}

# Создаем объект GridSearchCV для градиентного бустинга
gb_search = GridSearchCV(gb_classifier, gb_param_grid, cv=5)

# Запускаем поиск для градиентного бустинга
gb_search.fit(X_train, y_train)

# Выводим лучшие параметры и оценку для градиентного бустинга
print("Best parameters for Gradient Boosting:", gb_search.best_params_)
print("Accuracy of Gradient Boosting:", gb_search.best_score_)
```

## Нейронная сеть

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Определение архитектуры модели
def create_model(input_shape, num_classes):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=input_shape),
        layers.Dense(32, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Задаем размерности входных данных и количество классов
input_shape = (8,)  # 8 признаков
num_classes = 9     # 9 классов

# Создаем модель
model = create_model(input_shape, num_classes)

# Компилируем модель с функцией потерь sparse_categorical_crossentropy
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['sparse_categorical_accuracy'])

# Выводим информацию о модели
model.summary()
```

```python
from tqdm import tqdm

# Определяем количество эпох обучения
epochs = 200

# Создаем пустой список для сохранения истории метрик обучения
history = []

y_corr= y - 1

# Цикл обучения
for epoch in range(epochs):
    # Запускаем обучение модели и сохраняем историю обучения
    epoch_history = model.fit(X, y_corr, epochs=1, validation_split=0.2, verbose=0)
    # Добавляем метрики текущей эпохи в общую историю
    history.append(epoch_history.history)
    # Отображаем прогресс обучения с помощью tqdm
    tqdm.write(f"Epoch {epoch + 1}/{epochs}, Loss: {epoch_history.history['loss'][0]}, Accuracy: {epoch_history.history['sparse_categorical_accuracy'][0]}, Validation Loss: {epoch_history.history['val_loss'][0]}, Validation Accuracy: {epoch_history.history['val_sparse_categorical_accuracy'][0]}")
```

```python
# Получаем значения потерь и точности из истории обучения
train_loss = [epoch['loss'][0] for epoch in history]
val_loss = [epoch['val_loss'][0] for epoch in history]
train_accuracy = [epoch['sparse_categorical_accuracy'][0] for epoch in history]
val_accuracy = [epoch['val_sparse_categorical_accuracy'][0] for epoch in history]

# Нарисуем график потерь
plt.figure(figsize=(10, 5))
plt.plot(train_loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Нарисуем график точности
plt.figure(figsize=(10, 5))
plt.plot(train_accuracy, label='Train Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
